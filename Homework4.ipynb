**Cálculo de Pi con Monte Carlo**
\
Con este programa pretendemos a partir de Monte Carlo calcular el número pi, pero con el twist de colocar puntos aleatorios dentro de un cubo para estimar el volumen de una esfera. De esta menra queremos comprobar si las estimaciones mejoran frente al caso 2D o no.

```python
import numpy as tecno
import matplotlib.pyplot as graf
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
```

```python
#Establecemos el número total de puntos random
puntos_random_total=int(input("\nEscriba el número de puntos random que desea utilizar: "))

#Establecemos el número inicial de puntos
in_esfera=0
in_cubo=0

#Creamos seis listas vacías: tres para indicar las coordenadas de los puntos dentro del cuadrado, y otros tres para indicar los que estan dentro del círculo
x_cubo=[]
y_cubo=[]
z_cubo=[]

x_esfera=[]
y_esfera=[]
z_esfera=[]

#Establecemos las condiciones del cálculo
while(in_cubo < puntos_random_total):
  #Generamos puntos random dentro del espacio del cubo
  x=tecno.random.uniform(-1.0, 1.0)
  y=tecno.random.uniform(-1.0, 1.0)
  z=tecno.random.uniform(-1.0, 1.0)
  in_cubo=in_cubo+1

  #Comprobamos si estos puntos estan dentro de la esfera
  if x**2+y**2+z**2 <= 1.0:
    in_esfera=in_esfera+1
    x_esfera.append(x)
    y_esfera.append(y)
    z_esfera.append(z)
  else:
    x_cubo.append(x)
    y_cubo.append(y)
    z_cubo.append(z)

#Calculamos el ratio de puntos que han caido dentro de la esfera con respecto al total
ratio_esfera=in_esfera/in_cubo

#Conociendo el ratio y el volumen del cubo, podemos estimar el valor de pi
prox_pi=(ratio_esfera*8)*(3/4)

#Por último imprimimos todos los resultados y creamos las gráficas para observar el cubrimiento de estos puntos en el espacio
print('\nʕ•̫͡•ʕ•̫͡•ʔ•̫͡•ʔ•̫͡•ʕ•̫͡•ʔ•̫͡•ʕ•̫͡•ʕ•̫͡•ʔ•̫͡•ʔ•̫͡•ʕ•̫͡•ʔ•̫͡•ʔʕ•̫͡•ʕ•̫͡•ʔ•̫͡•ʔ•̫͡•ʕ•̫͡•ʔ•̫͡•ʕ•̫͡•ʕ•̫͡•ʔ•̫͡•ʔ•̫͡•ʕ•̫͡•ʔ•̫͡•ʔ')
print('\nResultados')
print('\nNúmero de puntos dentro del cubo de volumen 8:', in_cubo)
print('Número de puntos dentro de la esfera:',in_esfera,)
print('Ratio porcentual de puntos dentro de la esfera con respecto al cubo=',ratio_esfera*100,'% \n')
print('Volumen estimado de la esfera =',ratio_esfera,'x 8 x 3/4 =', prox_pi)
print('Exact value (pi):', tecno.pi)
print('**Estimación de Pi con MC**')

#Gráfica
figura = graf.figure()
xd = figura.add_subplot(111, projection='3d')
xd.scatter(x_cubo, y_cubo, z_cubo, c='b', marker='o')
xd.scatter(x_esfera, y_esfera, z_esfera, c='r', marker='o')

graf.show()
```
